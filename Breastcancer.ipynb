# **Breast Cancer Detection with neutral network**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

### Data collection and Processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
#print the first 5 rows of the dataframe
data_frame.head()
```

```python
#adding the target column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
data_frame.shape
```

```python
data_frame.info()
```

```python
data_frame.describe()
```

```python
#checking the distribution of target variable
data_frame['label'].value_counts()
```

1 -> Benign
0 -> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

### **Building the Neutral Network**

```python
#importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)   #to get same accuracy score
from tensorflow import keras   #deep learning library
```

```python
#setting up the layers of Neutral Network
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(128, input_shape=(30,), activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
#compiling the neutral network
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
#training the Neutral Network
history = model.fit(X_train_std, Y_train,validation_split = 0.1, epochs=100)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='upper right')
```

Accuracy of the model on test data

```python
loss,accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

model.predict()gives the prediction probability of each class for that data point

```python
#converting the prediction probability to class labels

Y_pred = model.predict(X_test_std)
Y_pred_labels = [1 if i > 0.5 else 0 for i in Y_pred]
print(Y_pred_labels)
```

### **Building the predictive system**

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.139,0.2479,0.885,15.34,0.006391,0.02058,0.02578,0.01242,0.01882,0.004434,15.11,19.26,99.76,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

#change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)
prediction = model.predict(input_data_std)
print(prediction)
prediction_label = [1 if i > 0.5 else 0 for i in prediction]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The breast cancer is Malignant')
else:
  print('The breast cancer is Benign')
```
